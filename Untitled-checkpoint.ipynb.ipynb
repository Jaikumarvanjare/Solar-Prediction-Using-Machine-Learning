```python
df=pd.read_csv("SolarPrediction.csv")
```

```python
df['Data']=df['Data'].apply(lambda x: x.split()[0])
```

```python
df['Month']=pd.to_datetime(df['Data']).dt.month
df['Day']=pd.to_datetime(df['Data']).dt.day
df['Year']=pd.to_datetime(df['Data']).dt.year
df['Hour']=pd.to_datetime(df['Time']).dt.hour
df['Minute']=pd.to_datetime(df['Time']).dt.minute
df['Second']=pd.to_datetime(df['Time']).dt.second
```

```python
df['SunriseHour']=df['TimeSunRise'].astype(str).apply(lambda x: re.search(r'^\d+',x).group(0) if re.search(r'^\d+',x) else np.nan).astype(float).astype('Int64')
df['SunriseMinute']=df['TimeSunRise'].astype(str).apply(lambda x: re.search(r':(\d+)',x).group(1) if re.search(r':(\d+)',x) else np.nan).astype(float).astype('Int64')
df['SunsetHour']=df['TimeSunSet'].astype(str).apply(lambda x: re.search(r':(\d+)',x).group(1) if re.search(r':(\d+)',x) else np.nan).astype(float).astype('Int64')
df['SunsetMinute']=df['TimeSunSet'].astype(str).apply(lambda x: re.search(r':(\d+)',x).group(1) if re.search(r':(\d+)',x) else np.nan).astype(float).astype('Int64')
```

```python
df.drop(['UNIXTime','Data','Time','TimeSunRise','TimeSunSet'],axis=1,inplace=True)
```

```python
input_features=df.drop(['Radiation'],axis=1)
output_features=df['Radiation']
```
